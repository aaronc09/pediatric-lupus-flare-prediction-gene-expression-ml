```python
# =========================================================
# Pediatric Lupus Flare Prediction (GSE65391)
# Author: Aaron Choi
#
# Colab runner for preprocessing pipeline
# Script: scripts/build_lupus_dataset.py
# =========================================================

# ====================================
# 1. MOUNT GOOGLE DRIVE
# ====================================
from google.colab import drive
drive.mount("/content/drive")

# ====================================
# 2. INSTALL DEPENDENCIES
# ====================================
!pip install -q GEOparse pyarrow pandas

# ====================================
# 3. CLONE GITHUB REPO (SAFE)
# ====================================
import os

REPO_PATH = "/content/pediatric-lupus-flare-prediction-gene-expression-ml"

if not os.path.exists(REPO_PATH):
    print("Cloning GitHub repo...")
    !git clone https://github.com/aaronc09/pediatric-lupus-flare-prediction-gene-expression-ml.git
else:
    print("Repo already exists. Skipping clone.")

# ====================================
# 4. IMPORTS
# ====================================
import sys
import json
import pandas as pd
import GEOparse

# ====================================
# 5. PATHS
# ====================================
SCRIPT_PATH = f"{REPO_PATH}/scripts/build_lupus_dataset.py"
OUTPUT_DIR = "/content/drive/MyDrive/lupus_project_outputs"

# ====================================
# 6. VALIDATE PATHS
# ====================================
if not os.path.exists(REPO_PATH):
    raise FileNotFoundError(f"Repo path does not exist: {REPO_PATH}")

if not os.path.exists(SCRIPT_PATH):
    raise FileNotFoundError(f"Script file does not exist: {SCRIPT_PATH}")

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ====================================
# 7. MOVE INTO REPO FOLDER
# ====================================
%cd $REPO_PATH

# ====================================
# 8. RUN PREPROCESSING SCRIPT
# ====================================
print("\nRunning preprocessing pipeline...\n")
!python "$SCRIPT_PATH" --out "$OUTPUT_DIR"

# ====================================
# 9. LIST GENERATED FILES
# ====================================
print("\nSaved files:")
!ls -lh "$OUTPUT_DIR"

# ====================================
# 10. LOAD AND PRINT PIPELINE SUMMARY
# ====================================
summary_file = f"{OUTPUT_DIR}/pipeline_summary.json"

if not os.path.exists(summary_file):
    raise FileNotFoundError(f"Missing summary file: {summary_file}")

with open(summary_file, "r") as f:
    summary = json.load(f)

print("\nPipeline summary:")
for k, v in summary.items():
    print(f"{k}: {v}")

# ====================================
# 11. LOAD FINAL DATASET
# ====================================
final_df_path = f"{OUTPUT_DIR}/lupus_final_df.pkl"

if not os.path.exists(final_df_path):
    raise FileNotFoundError(f"Missing final dataset file: {final_df_path}")

df = pd.read_pickle(final_df_path)

# ====================================
# 12. QUICK SANITY CHECKS
# ====================================
print("\nDataset shape:", df.shape)
print("Unique subjects:", df["subject"].nunique())

print("\nPreflare distribution:")
print(df["preflare_bool"].value_counts())

print("\nFirst 5 rows of final dataset:")
display(df.head())

# ====================================
# 13. ENVIRONMENT INFO
# ====================================
print("\nColab environment:")
print("Python:", sys.version)
print("pandas:", pd.__version__)
print("GEOparse:", GEOparse.__version__)
```
